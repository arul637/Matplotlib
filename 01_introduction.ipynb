```python
# importing the modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
np.random.seed(seed=7)

x = np.arange(1, 10+1)
y = np.random.randint(25, 38, size=10)
```

```python
weather_data = {
    'Days': x,
    'Temperature': y
}
df = pd.DataFrame(weather_data)
df
```

```python
plt.title('Weather Data')
plt.xlabel('Days')
plt.ylabel('Temperature')
plt.plot(x, y, linewidth=1.5, color='red', label='weather report')
plt.legend()

plt.show()
```
