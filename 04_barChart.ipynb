```python
# importing the modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
company = ['GOOGLE','AMAZON','MICROSOFT','FACEBOOK']
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

```python
US_stock_report = {
    'company': company,
    'revenue': revenue,
    'profit': profit
}

df = pd.DataFrame(US_stock_report)
df
```

```python
xpos = np.arange(4)
xpos
```

```python
plt.title('US stock reports')

plt.bar(xpos, revenue)
plt.show()
```

to change the x axis as to labels, using the xtick function()

```python
plt.xticks(xpos, company)
plt.title('US stock reports')
plt.ylabel('revenue and profits')
plt.bar(xpos, revenue)
plt.show()
```

### inserting the profit bars

```python
plt.xticks(xpos, company)
plt.title('US stock reports')
plt.ylabel('revenue and profits')
plt.bar(xpos, revenue)
plt.bar(xpos, profit)
plt.show()
```

```python
plt.xticks(xpos, company)
plt.title('US stock reports')
plt.ylabel('revenue and profits')
plt.bar(xpos-0.2, revenue, label='revenue', width=0.4)
plt.bar(xpos+0.2, profit, label='profit', width=0.4)
plt.legend()
plt.show()
```

## horizontal bars

```python
plt.yticks(xpos, company)
plt.title('US stock reports')
plt.xlabel('revenue and profits')
plt.barh(xpos-0.2, revenue, height=0.4, label='revenue')   # barh means bar chart in horizontal
plt.barh(xpos+0.2, profit, height=0.4, label='profit')     #            ___          _
plt.legend()
plt.show()
```
