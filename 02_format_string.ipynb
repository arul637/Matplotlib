```python
# importing the modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
np.random.seed(seed=7)                      # for random in static

x = np.arange(1, 10+1)
y = np.random.randint(25, 35, size=10)
```

```python
weather_data = {
    'days': x,
    'temperature': y
}

df = pd.DataFrame(weather_data)
df
```

### plotting without string format

```python
plt.title('weather data')
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(x, y, color='green', linewidth=1.5, linestyle='dashed', marker='D')
plt.show()
```

```python
plt.title('weather data')
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(x, y, color='violet', linewidth=1.5, linestyle='dotted', marker='o')
plt.show()
```

```python
plt.title('weather data')
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(x, y, color='red',alpha=0.6, linewidth=1.5, marker='o', label='weather report')    # legend will show the report
plt.legend()
plt.show()
```

```python
plt.title('weather data')
plt.xlabel('days')
plt.ylabel('temperature')
plt.plot(x, y, color='green', alpha=0.5 ,linewidth=1.5)
plt.plot(x, y, color='red', marker='^', linewidth=0)
plt.show()
```
