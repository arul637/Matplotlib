```python
# importing the modules
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# creating the datas
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
blood_sugar_level = {
    'mens': blood_sugar_men,
    'womens': blood_sugar_women
}

df = pd.DataFrame(blood_sugar_level)
df
```

### plotting the simple histogram

```python
plt.hist(blood_sugar)
```

### plotting with the perfect bins parameter

by default it tables 10 bins
(array([3., 3., 1., 0., 1., 1., 0., 2., 0., 2.]),
 array([ 77. ,  84.3,  91.6,  98.9, 106.2, 113.5, 120.8, 128.1, 135.4,
        142.7, 150. ]),
 
 the second array considered as bins

```python
plt.hist(blood_sugar, bins=3)
```

we need little gap between each histogram

```python
plt.hist(blood_sugar, bins=3, rwidth=0.8) # rwidth [relative width] 
```

#### Histogram showing normal, pre-diabetic and diabetic patients distribution

   - 80-100: Normal
   - 100-125: Pre-diabetic
   - 80-100: Diabetic

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 150], color='green', rwidth=0.95)
```

### multiple datasamples in histogram

```python
plt.title('blood sugar levels')

plt.ylabel('no of patients')
plt.xlabel('diabetics level')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'])

plt.legend()
```

```python
plt.title('blood sugar levels')

plt.ylabel('no of patients')
plt.xlabel('diabetics level')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'], histtype='step')
plt.legend()
```

# horizontal orientation

```python
plt.title('blood sugar levels')

plt.ylabel('no of patients')
plt.xlabel('diabetics level')

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green', 'orange'],
         label=['men', 'women'], orientation='horizontal')
plt.legend()
```
