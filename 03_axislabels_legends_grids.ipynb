```python
# importing the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# creating the datas

np.random.seed(seed=7)                          # constant random at all time

days = np.array([1, 2, 3, 4, 5, 6, 7])
max_temperature = np.random.randint(35, 55, size=7)
min_temperature = np.random.randint(25, 35, size=7)
avg_temperature = [np.sum([max_temperature[i], min_temperature[i]])//2 for i in range(len(days))]
```

```python
weather_data = {
    'days': days,
    'maximum temperature': max_temperature,
    'minimum temperature': min_temperature,
    'average temperature': avg_temperature
}

df = pd.DataFrame(weather_data)
df
```

### plotting the simpe graph using axis labels

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather report")
plt.plot(days, max_temperature)
plt.plot(days, min_temperature)
plt.plot(days, avg_temperature)

plt.show()
```

### plotting the legends

```python
plt.xlabel("days")
plt.ylabel("temperature")
plt.title("weather report")
plt.plot(days, max_temperature, label='max')
plt.plot(days, min_temperature, label='min')
plt.plot(days, avg_temperature, label='avg')

plt.legend(loc='best', fontsize='small')

plt.show()
```
