```python
# importing the modules
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# generting the labels
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

### creating the simle pie chart

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.pie(exp_vals, labels=exp_labels)
plt.show()
```

### showing the percentage 

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%')
plt.show()
```

### changing the radius of the pie

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5)
plt.show()
```

### shadow in pie chart

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5, shadow=True)
plt.show()
```

### removing the 1 pie in it

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5, shadow=False, explode=[0, 0, 0, 0.2, 0])
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5, shadow=False, explode=[0, 0.1, 0, 0, 0])
plt.show()
```

### rotating the pie

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5, shadow=False, explode=[0, 0, 0, 0, 0], startangle=45)
plt.show()
```

```python
plt.pie(exp_vals, labels=exp_labels, autopct='%0.2f%%', radius=1.5, shadow=False, explode=[0, 0.15, 0, 0.1, 0], startangle=180)
plt.show()
```
